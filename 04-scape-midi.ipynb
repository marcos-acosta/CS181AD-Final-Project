## 4. Scrape MIDI files

This notebook scrapes MIDI files from [https://bushgrafts.com/midi/](https://bushgrafts.com/midi/) and saves them to a directory. It's not necessary to run this notebook, as processed versions are available in later notebooks.

```python
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
MIDI_SAVEDIR = 'midi/'
```

```python
link = "https://bushgrafts.com/midi/"
f = requests.get(link)
soup = BeautifulSoup(f.content, "html.parser")
```

```python
links_div = soup.find("div", class_="entry-content-inner")
links = links_div.find_all("a", href=True)[29:]
```

```python
links
```

```python
links = [link['href'] for link in links if link.text != 'Return to top']
```

```python
links
```

```python
for link in tqdm(links):
    name = link.split('/')[-1].split('.')[0]
    name = name.replace('%20', '-')
    r = requests.get(link, allow_redirects=True)
    open(f'{MIDI_SAVEDIR}{name}.mid', 'wb').write(r.content)
```
